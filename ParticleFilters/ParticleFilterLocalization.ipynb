### Some background on localization

The central question is: Where am I?

Localization is the problem of finding out a robot’s coordinates relative to its environment, assuming that one is provided with a map of the environment. Localization is a key component in various successful mobile robot systems. Occasionally is has been referred to as “ the most fundamental problem of providing a mobile robot with autonomous capabilities”. Particularly challenging is the global localization problem, where the robot does not know its initial position and therefore has to globally localize itself (See [2]). 

The goal of the localization is to pinpoint location and pose of the robot in the global coordinate system using least amount of resources and time. Although we have actuation commands, linear encoders, and sensors that can be used combinedly to achieve this goal, the characteristic of any sophisticated robotic application is that they operate in increasingly unstructured environments, environments that are inherently unpredictable. Uncertainty arises from various factors such as: 1) Environments: dynamic/unpredictable 2) Sensors: range/resolution/noise 3) Motion/Control: inaccurate/wear and tear 4) Models: inherently inaccurate 5) Computation: accuracy. 

### Probabilistic state estimation

The core problem is to estimate state from the observable data. The state is e.g., global location and pose can not be directly observable, but it can be inferred through sensor measurements and control commands. However, with the uncertainty arising from various factors as mentioned earlier, the state cannot be estimated precisely. To deal with the ambiguities involved, Probabilistic state estimation algorithms represent state by a probability distribution over a whole space of possible world states.

In this experiment, we have a robot object whose motion and heading direction can be controlled programmatically. We will only use these parameters (location/pose) to simulate its motion in 2D and for the visualization purpose only. To simulate practical situation, we assume that we do not have direct access to reading these parameters. The robot is equipped with sensors, and we can only observe the sensor measurements as well as motion commands. We model uncertainty associated with these as a gaussian distribution with zero mean. Using this noisy data, the localization algorithm has to infer accurate location and pose of the robot. 

In every step, the robot makes a movement, senses the environment to make measurements (distance from landmarks), and repeats this cycle. During motion, we lose the information because of the uncertainty associated with the motion and inaccuracies of physical models, whereas any sensor measurement gains new information. Although measurements are noisy, we still get some information about where we are in the world and hence they reduce overall uncertainty of the state.

## Environment

````python
# Environment parameters
g_environment = { 
    # "landmarks": [[30.0, 30.0], [170.0, 70.0], [40.0, 170.0], [160.0, 130.0]], 
    "landmarks": [[30.0, 30.0], [170.0, 30.0], [30.0, 170.0], [170.0, 170.0]], 
    "world_size": 200.0 
}

sensor_noise = 30.0 # this is quite large compared to the world size
turning_noise, forward_noise = 0.08, 1.0
````

### Gaussian Noise

Let's assume that our world is 200x200 grid, and are total four landmarks to use as a reference. The landmark locations are:

```
"landmarks": [[30.0, 30.0], [170.0, 30.0], [30.0, 170.0], [170.0, 170.0]]
```

Let's further assume our robot is located at the center of the grid i.e., (100,100). The precise measurement from the landmarks would be: 

```
z = [[-70., -70.], [ 70., -70.], [-70.,  70.], [ 70.,  70.]]
```

Now, we want to model noise/uncertainty associated with it. We simply, add a gaussian random variable to each of the observable quantity. The sampling distribution of this random variable has zero mean and standard deviation approximating uncertainty that we want to model. We assume that the sensor noise has a standard deviation of 30. 

```
Noise N(μ, σ) with μ = 0.0 and σ = 20.0. 
=> Pr(μ - 3σ ≤ noise ≤ μ + 3σ) ≈ 0.9772 i.e. Pr(-90 ≤ noise ≤ 90) ≈ 0.9772
```

So, observation with uncertainty would be:

```
z_sensor = z + N(μ, σ)
```

Let's visualize the effects of this uncertainty.

````python
np.random.seed(2) # for consistent results
````

````python
robot_pos = (100,100)
z = np.array([[-70., -70.], [ 70., -70.], [-70.,  70.], [ 70.,  70.]])
z_sensor = z + np.random.normal(0.0, 30.0, (4,2))
print "True observation: \n", z
print "Noisy observation: \n", z_sensor

plt.plot(robot_pos[0], robot_pos[1], "kx", markersize=10, label="robot")

lm = np.array(g_environment["landmarks"])
plt.plot(lm[:, 0], lm[:, 1], c="y", marker="o", markersize=10, linestyle = 'None', label="true observation")
plt.plot(lm[:, 0], lm[:, 1], c="b", marker="*", markersize=10, linestyle = 'None')
plt.plot(robot_pos[0]+z_sensor[:,0], robot_pos[1]+z_sensor[:,1], "ro", label="noisy observation")
plt.legend(loc='best')
plt.tight_layout()
````

We see that measurements are very inaccurate compared to the world size our robot moves in. 

````python
# Robot class
class Robot:
    '''
    x, y: 2d world locations
    orientation: robot orientation in radians
    forward_noise: uncertainty in forward motion
    turn_noise: uncertainty in turning
    sense_noise: uncertainty in sensor readings
    '''
    def __init__(self):
        self.x = np.random.random() * g_environment["world_size"]
        self.y = np.random.random() * g_environment["world_size"]
        self.orientation = np.random.random() * 2.0 * np.pi
        self.forward_noise = 1e-100;
        self.turn_noise    = 1e-100;
        self.sense_noise   = 1e-100;
    
    def set(self, new_x, new_y, new_orientation):
        if new_x < 0 or new_x >= g_environment["world_size"]:
            raise ValueError, 'X coordinate out of bound'
        if new_y < 0 or new_y >= g_environment["world_size"]:
            raise ValueError, 'Y coordinate out of bound'
        if new_orientation < 0 or new_orientation >= 2 * np.pi:
            raise ValueError, 'Orientation must be in [0..2np.pi]'
        self.x = float(new_x)
        self.y = float(new_y)
        self.orientation = float(new_orientation)
    
    def set_noise(self, new_f_noise, new_t_noise, new_s_noise):
        # makes it possible to change the noise parameters
        # this is often useful in particle filters
        self.forward_noise = float(new_f_noise);
        self.turn_noise    = float(new_t_noise);
        self.sense_noise   = float(new_s_noise);
    
    def sense(self):
        Z = []
        for i in range(len(g_environment["landmarks"])):
            dist = np.sqrt((self.x - g_environment["landmarks"][i][0]) ** 2 \
                            + (self.y - g_environment["landmarks"][i][1]) ** 2)
            dist += np.random.normal(0.0, self.sense_noise)
            Z.append(dist)
        return Z
    
    def move(self, turn, forward):
        if forward < 0:
            raise ValueError, 'Robot cant move backwards'         
        
        # turn, and add randomness to the turning command
        orientation = self.orientation + float(turn) + np.random.normal(0.0, self.turn_noise)
        orientation %= 2 * np.pi
        
        # move, and add randomness to the motion command
        dist = float(forward) + np.random.normal(0.0, self.forward_noise)
        x = self.x + (np.cos(orientation) * dist)
        y = self.y + (np.sin(orientation) * dist)
        x %= g_environment["world_size"]    # cyclic truncate
        y %= g_environment["world_size"]
        
        # set particle
        res = Robot()
        res.set(x, y, orientation)
        res.set_noise(self.forward_noise, self.turn_noise, self.sense_noise)
        return res
    
    def Gaussian(self, mu, sigma, x):
        
        # calculates the probability of x for 1-dim Gaussian with mean mu and var. sigma
        return np.exp(- ((mu - x) ** 2) / (sigma ** 2) / 2.0) / np.sqrt(2.0 * np.pi * (sigma ** 2))
      
    def measurement_prob(self, measurement):
        
        # calculates how likely a measurement should be
        
        prob = 1.0;
        for i in range(len(g_environment["landmarks"])):
            dist = np.sqrt((self.x - g_environment["landmarks"][i][0]) ** 2 \
                               + (self.y - g_environment["landmarks"][i][1]) ** 2)
            prob *= self.Gaussian(dist, self.sense_noise, measurement[i])
        return prob
      
    def __repr__(self):
        return '[x=%.6s y=%.6s orient=%.6s]' % (str(self.x), str(self.y), str(self.orientation))

def eval(r, p):
    sum = 0.0;
    world_size = g_environment["world_size"]
    for i in range(len(p)): # calculate mean error
        dx = (p[i].x - r.x + (world_size/2.0)) % world_size - (world_size/2.0)
        dy = (p[i].y - r.y + (world_size/2.0)) % world_size - (world_size/2.0)
        err = np.sqrt(dx * dx + dy * dy)
        sum += err
    return sum / float(len(p))
````

````python
# plotting utility 
def plot_curr_state(environment, robot_t, particles_t=None):
    
    # landmarks
    for lm in environment["landmarks"]:
        plt.plot(lm[0], lm[1], c="y", marker="o", zorder=0, markersize=20)
        plt.plot(lm[0], lm[1], c="b", marker="*", zorder=0, markersize=20)
    
    # particles
    if particles_t:
        for p in particles_t:
            plt.plot(p.x, p.y, marker=(4, 0, p.orientation*180/np.pi+45), 
                                             c="skyblue", alpha=0.3, zorder=1, markersize=15)
            plt.plot(p.x, p.y, marker=(3, 0, p.orientation*180/np.pi-90), 
                                             c="steelblue", alpha=0.3, zorder=1, markersize=10)
        
        # particle estimation
        mu_x, mu_y = np.array([(p.x, p.y, p.orientation) for p in particles])[:,0:2].mean(axis=0)
        plt.plot(mu_x, mu_y, c="steelblue", alpha=1.0, \
                                 marker=(4, 0, robot_t.orientation*180/np.pi+45), zorder=2, markersize=25)
        plt.plot(mu_x, mu_y, c="yellow", alpha=1.0, \
                                 marker=(3, 0, robot_t.orientation*180/np.pi-90), zorder=2, markersize=22)
        plt.plot(mu_x, mu_y, c="k", marker='.', zorder=2)
    
    # robot
    if robot_t:
        plt.plot(robot_t.x, robot_t.y, c="g", alpha=1.0,
                                marker=(4, 0, 45+robot_t.orientation*180/np.pi), zorder=5, markersize=30)
        plt.plot(robot_t.x, robot_t.y, c="lime", alpha=1.0, 
                                marker=(3, 0, robot_t.orientation*180/np.pi-90), zorder=5, markersize=30)
        plt.plot(robot_t.x, robot_t.y, c="k", marker='.', zorder=7)

    plt.xlim([0, environment["world_size"]])
    plt.ylim([0, environment["world_size"]])
    plt.tight_layout()
    plt.rcParams.update({'font.size':16})
````

## Particle Filter

Each particle in Particle Filter represents a probable location and pose of the target robot. During initialization, we distribute a constant (N) amount of particles uniformly at random. This amounts to the complete uncertainty about robot location in the beginning.

As the robot makes its first observation, we compute measurement probability for each particle. For example, if the robot started at the center, then it is less likely that particles sitting at the right end would make similar observations (given that measurement uncertainty is not crazily high). You can imagine further steps as a survival game in which only those particles get to survive which closely represent our robot's location and pose. However, doing so will decrease its population over time. Therefore, we don't kill the rest of the particles. Instead, we resample the same amount of particles, however, at this time, they aren't sampled uniformly at random. We now have gained some knowledge about robot location and pose (by measuring measurement probability for each particle). We use that knowledge to define our new sampling distribution. The benefits are twofold: to maintain a constant amount of particles as well as to improve precision as we progress. 

Before we resample, we assign weights to all the particles depending on how closely they are related to the actual robot's location and pose. Then during resampling, we sample particles according to their weights. The higher the weight, the more chance of being sampled. We use Resampling Wheel Algorithm for this purpose. After few resampling, we can clearly see that the new particle's density is directly proportional to the measurement probability, so we see more particles around the actual robot then away from it. 

In the visualization, the green object represents our robot and the yellow object is the best estimation of its location calculated using particle filter algorithm. 

### Initialization

````python
# Create particles
N = 400
particles = []
for i in range(N):
    r = Robot()
    # set uncertainty in motion (turn/forward) and sensor readings
    r.set_noise(forward_noise, turning_noise, sensor_noise) 
    particles.append(r)
````

````python
plot_curr_state(g_environment, myrobot, particles)
plt.title("Initial State with Particles")
````

### Move-Sense loop

````python
# We assume there is no external force involved, and there is a constant motion
motion = (np.random.random()*1.0, 5.0) # (turn, forward) 

eval_hist = []
iter_no = 0
rmse = eval(myrobot,particles)

fig = plt.figure(figsize=(8,6))
for i in range(30):
    
    iter_no += 1
    
    # Plot current state of the environment, our robot, and particles
    plt.gca().cla()
    plot_curr_state(g_environment, myrobot, particles)
    plt.title("Iter: {}, RMSE: {}".format(iter_no, round(rmse,4)))
    display.clear_output(wait=True)
    display.display(plt.gcf())
    
    # Move (move the actual robot) - increases uncertainty
    myrobot = myrobot.move(*motion)
    
    # Measurement (here, we will measure the actual distance of the robot from all the landmarks)
    # - reduces uncertainty
    Z = myrobot.sense()
    
    p2 = []
    for i in range(N):
        # For each particle, we apply the same motion 
        # Remember that these particles will have uncertainty in the motion
        p2.append(particles[i].move(*motion))
    particles = p2
    
    w = []
    for i in range(N):
        # For each particle, we calculate how likely is the measurement that we got from actual 
        # robot given the particle's location (we ignore the orientation for the simplicity)
        # For example, if the robot is at the center of the world, then it is less likely that 
        # a particle sitting at the corner would make the same measurement (assuming our measurement 
        # uncertainty is not crazily high)
        # We store this probability as weights. The closer the particle to the actual robot, 
        # the higher the weight it will be assigned.
        w.append(particles[i].measurement_prob(Z))

    # Resampling wheel algorithm 
    # We will resample particles according to their weights in order to get partilces
    # which are more likely to be in proximity of our target robot in both spatial and pose dimension.
    p3 = []
    index = int(np.random.random() * N)
    beta = 0.0
    mw = max(w)
    for i in range(N):
        beta += np.random.random() * 2.0 * mw
        while beta > w[index]:
            beta -= w[index]
            index = (index + 1) % N
            
        p3.append(particles[index])
    particles = p3
    
    rmse = eval(myrobot,particles)
    print "RMSE: ", rmse
    eval_hist.append(rmse)
    
    fig.savefig("pf{}.png".format(iter_no), dpi=fig.dpi)
    # x_dev, y_dev = np.array([(p.x, p.y, p.orientation) for p in particles])[:,0:2].std(axis=0)
    # particle_dev = x_dev*y_dev
    # time.sleep(0.5)
plt.clf()
# print "RMSE log: ", eval_hist
````
